Questão 5

```python
dignidade = int(input("Qual seu nível de dignidade? (0-100): "))
if dignidade >= 90 and dignidade <= 100:
  print("Você levantou o martelo!")
elif dignidade >= 0 and dignidade < 90:
  print("Você é indigno.")
else:
  print("Valor inválido.")
```

```python
distanciaTeia = int(input("Qual a distância de disparo? "))

if distanciaTeia < 10 and distanciaTeia > 0:
  print("Disparo Curto")
elif distanciaTeia >= 10 and distanciaTeia <= 50:
  print("Disparo Médio")
elif distanciaTeia >= 50:
  print("Disparo Longo")
else:
  print("Valor Inválido")
```
