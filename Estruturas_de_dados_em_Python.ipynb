MISSÃO 15: Dicionário de PreçosNível: Agente

```python
precos_armas = {
    "Vandal": 2900,
    "Phantom": 2900,
    "Spectre": 1600,
    "Ghost": 500,
    "Stinger": 1100
}

armas_baratas = []
for nome, preco in precos_armas.items():
    if preco < 2000:
        armas_baratas.append(nome)

print(armas_baratas)
```

MISSÃO 16: Analista de DesempenhoNível: Radiante

```python
jogadores = [
    {"nome": "Raze", "kills": 25, "deaths": 10},
    {"nome": "Cypher", "kills": 12, "deaths": 15},
    {"nome": "Jett", "kills": 30, "deaths": 5}
]

bons_jogadores = [p["nome"] for p in jogadores if p["kills"] > p["deaths"]]

print(bons_jogadores)
```

MISSÃO 20: Sistema de MVP FinalNível: Radiante

```python
jogadores = [
    {"nome": "Raze", "kills": 25, "deaths": 10},
    {"nome": "Cypher", "kills": 12, "deaths": 15},
    {"nome": "Jett", "kills": 30, "deaths": 5}
]

mvp = max(jogadores, key=lambda p: p["kills"] - p["deaths"])

saldo_mvp = mvp["kills"] - mvp["deaths"]

print(f"O MVP da partida é {mvp['nome']} com saldo de {saldo_mvp} kills")
```
